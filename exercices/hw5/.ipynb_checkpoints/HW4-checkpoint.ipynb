```python
import numpy as np
from numpy import exp, log, sqrt, tanh, sinh, cosh, sin, cos
import matplotlib.pyplot as plt
import scipy
from scipy.optimize import fsolve, fminbound
from scipy.integrate import quad
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
```

```python
# Problem 1(b)
# Binary estimation problem
def Phi_RS(m, lam):
    def integrand(z):
        cfun = cosh if m >= 0 else cos
        fun = log((cosh(2*lam*m) + cfun(2*sqrt(lam*abs(m))*z))/2)
        gaussian = exp(-z**2/2) / sqrt(2*np.pi)
        return fun * gaussian
    return -lam*m**2 / 4 - lam*m / 2 + quad(integrand, -30, 30)[0] / 2

plt.figure(figsize=(14,5))
lam_list = [0.5, 0.75, 1, 1.05, 1.1]
plt.subplot(1,2,1)
m = np.linspace(0,1,101)
for lam in lam_list:
    plt.plot(m, np.vectorize(Phi_RS)(m, lam), label=r'$\lambda = {0}$'.format(lam))
plt.legend()
plt.xlabel(r'$ m $')
plt.ylabel(r'$ \Phi_{\mathrm{RS}}(m;\lambda) $')

plt.subplots_adjust(wspace=0.3)
plt.subplot(1,2,2)
m = np.linspace(0,0.1,51)
for lam in lam_list:
    plt.plot(m, np.vectorize(Phi_RS)(m, lam), label=r'$\lambda = {0}$'.format(lam))
plt.legend()
plt.xlabel(r'$ m $')
plt.ylabel(r'$ \Phi_{\mathrm{RS}}(m;\lambda) $')

plt.savefig('hw4_1(b)1.pdf', bbox_inches='tight')
```

```python
def f(m, lam):
    def integrand(z):
        if m > 0:
            num = 2*sinh(2*lam*m) + z / sqrt(lam*m) * sinh(2*z*sqrt(lam*m))
            den = cosh(2*lam*m) + cosh(2*sqrt(lam*m)*z)
        else:
            num = 2*z**2
            den = 2
        gaussian = exp(-z**2/2) / sqrt(2*np.pi)
        return gaussian * num / den
    return -m - 1 + quad(integrand, -30, 30)[0]
def m_star(lam):
    return fsolve(f, 1, args=lam) if lam > 0 else 0
lam = np.linspace(0,10,101)
ms = np.fromiter((m_star(x) for x in lam), float)
mmse = 1 - ms
plt.figure(figsize=(14,5))
plt.subplot(1,2,1)
plt.plot(lam, ms)
plt.xlabel(r'$ \lambda $')
plt.ylabel(r'$ m^*(\lambda) $')
plt.subplot(1,2,2)
plt.plot(lam, mmse)
plt.xlabel(r'$ \lambda $')
plt.ylabel(r'$ \mathsf{MMSE}(\lambda) $')

plt.savefig('hw4_1(b)2.pdf', bbox_inches='tight')
```

```python
# Problem 3
# Problem 1(b)
# Binary estimation problem
def Phi_RS_bias(m, lam, rho):
    def integrand(z):
        lm, slm, lr, slr = lam * m, sqrt(lam * m), rho/(1-rho), sqrt(rho/(1-rho))
        
        Epp = exp( lm/2/lr + slm/slr*z )
        Epm = exp( -lm/2*(lr+2) - slr*slm*z )
        Emp = exp( -lm/2*(1/lr+2) + slm/slr*z )
        Emm = exp( lm/2*lr - slm*slr*z )
        
        T1_den = rho * Epp + (1-rho) * Epm
        T2_den = rho * Emp + (1-rho) * Emm
        
        gaussian = exp(-z**2/2) / sqrt(2*np.pi)
        return ( rho*log(T1_den) + (1-rho)*log(T2_den) ) * gaussian
    return -lam*m**2 / 4 + quad(integrand, -30, 30)[0]


plt.figure(figsize=(14,5))
rho = 0.075
lam_list = [0.5, 0.75, 1, 1.05, 1.1]
plt.subplot(1,2,1)
m = np.linspace(0,1,101)
for lam in lam_list:
    plt.plot(m, np.vectorize(Phi_RS_bias)(m, lam, rho), label=r'$\lambda = {0}$'.format(lam))
plt.legend()
plt.xlabel(r'$ m $')
plt.ylabel(r'$ \Phi_{\mathrm{RS}}(m;\lambda) $')

plt.subplots_adjust(wspace=0.3)
plt.subplot(1,2,2)
m = np.linspace(0,0.1,51)
for lam in lam_list:
    plt.plot(m, np.vectorize(Phi_RS_bias)(m, lam, rho), label=r'$\lambda = {0}$'.format(lam))
plt.legend()
plt.xlabel(r'$ m $')
plt.ylabel(r'$ \Phi_{\mathrm{RS}}(m;\lambda) $')

plt.savefig('hw4_3(a)1.pdf', bbox_inches='tight')
```

```python
def f(m, lam, rho):
    # m < 0 makes no sense, return a non-zero value indicating it is not the root
    if m < 0: return -1
    def integrand(z):
        lm, slm, lr, slr = lam * m, sqrt(lam * m), rho/(1-rho), sqrt(rho/(1-rho))
        S = z/2 * sqrt( lam * rho * (1-rho) / m )
        
#         Epp = exp( lm/2/lr + slm/slr*z )
#         Epm = exp( -lm/2*(lr+2) - slr*slm*z )
#         Emp = exp( -lm/2*(1/lr+2) + slm/slr*z )
#         Emm = exp( lm/2*lr - slm*slr*z )
        
        # Always cancel the larger exponential term for numerical stability
        cpp = lm/2/lr + slm/slr*z
        cpm = -lm/2*(lr+2) - slr*slm*z
        cmp = -lm/2*(1/lr+2) + slm/slr*z
        cmm = lm/2*lr - slm*slr*z
        
        Epp, Epm = (1, exp(cpm-cpp)) if cpp > cpm else (exp(cpp-cpm), 1)
        Emp, Emm = (1, exp(cmm-cmp)) if cmp > cmm else (exp(cmp-cmm), 1)
        
        T1_num = ( lam*(1-rho)/2 + S ) * Epp + ( -lam*(2-rho)/2 - S ) * Epm
        T1_den = rho * Epp + (1-rho) * Epm
        T1 = T1_num / T1_den
        T2_num = ( -lam*(1+rho)/2 + S ) * Emp + ( lam*rho/2 - S ) * Emm
        T2_den = rho * Emp + (1-rho) * Emm
        T2 = T2_num / T2_den
        
        gaussian = exp(-z**2/2) / sqrt(2*np.pi)
        return (rho*T1 + (1-rho)*T2) * gaussian
    return -lam*m/2 + quad(integrand, -30, 30)[0]
def m_star(lam, rho):
    return fsolve(f, 10, args=(lam, rho)) if lam > 0 else 0
rho = 0.075
lam = np.linspace(0,10,101)
ms = np.fromiter((m_star(x, rho) for x in lam), float)
mmse = 1 - ms
plt.figure(figsize=(14,5))
plt.subplot(1,2,1)
plt.plot(lam, ms)
plt.xlabel(r'$ \lambda $')
plt.ylabel(r'$ m^*(\lambda) $')
plt.subplot(1,2,2)
plt.plot(lam, mmse)
plt.xlabel(r'$ \lambda $')
plt.ylabel(r'$ \mathsf{MMSE}(\lambda) $')

plt.savefig('hw4_3(a)2.pdf', bbox_inches='tight')
```

```python
def Phi_RS(m, Delta, rho):
    lam = 1 / Delta
    def integrand(z):
        lm, slm, lr, slr = lam * m, sqrt(lam * m), rho/(1-rho), sqrt(rho/(1-rho))
        
#         cpp = lm/2/lr + slm/slr*z
#         cpm = -lm/2*(lr+2) - slr*slm*z
#         cmp = -lm/2*(1/lr+2) + slm/slr*z
#         cmm = lm/2*lr - slm*slr*z
        
#         Epp, Epm = (1, exp(cpm-cpp)) if cpp > cpm else (exp(cpp-cpm), 1)
#         Emp, Emm = (1, exp(cmm-cmp)) if cmp > cmm else (exp(cmp-cmm), 1)
        
        Epp = exp( lm/2/lr + slm/slr*z )
        Epm = exp( -lm/2*(lr+2) - slr*slm*z )
        Emp = exp( -lm/2*(1/lr+2) + slm/slr*z )
        Emm = exp( lm/2*lr - slm*slr*z )
        
        T1_den = rho * Epp + (1-rho) * Epm
        T2_den = rho * Emp + (1-rho) * Emm
        
        gaussian = exp(-z**2/2) / sqrt(2*np.pi)
        return ( rho*log(T1_den) + (1-rho)*log(T2_den) ) * gaussian
    return -lam*m**2 / 4 + quad(integrand, -30, 30)[0]

def f_in(A, B, rho):
    T1 = exp(-A/2*(1-rho)/rho + B*sqrt((1-rho)/rho))
    T2 = exp(-A/2*rho/(1-rho) - B*sqrt(rho/(1-rho)))
    return sqrt(rho*(1-rho)) * (T1-T2) / ( rho*T1 + (1-rho)*T2 )
    
def f_SE(m, Delta, rho):
    x, r = m / Delta, rho * (1-rho)
    def integrand(z):
        w = x/r/2 + z*sqrt(x/r)
        gaussian = exp(-z**2/2) / sqrt(2*np.pi)
        temp = 2*r*sinh(w) / (1 + 2*r*(cosh(w)-1))
        return gaussian * temp
    return quad(integrand, -30, 30)[0]

def find_fixed_points(Delta, rho, max_iter=1000, tol=1e-8):
    m_in, m_un = 1, 1e-5
    for i_un in range(max_iter):
        m_un_new = f_SE(m_un, Delta, rho)
        if abs(m_un_new-m_un) < tol: break
        m_un = m_un_new
    for i_in in range(max_iter):
        m_in_new = f_SE(m_in, Delta, rho)
        if abs(m_in_new-m_in) < tol: break
        m_in = m_in_new
    m_unstable = fminbound(lambda m: Phi_RS(m, Delta, rho), m_un, max(m_in, m_un))
    return m_un, m_in, m_unstable

def find_Delta_threshold(bound, rho, max_iter=1000, tol=1e-8, precision=1e-4):
    left, mid, right = bound
    # Find Delta_Alg in [left, mid]
    low, high = left, mid
    while high - low > 1e-4:
        mid = (low + high) / 2
        Delta = mid * rho**2
        m_un, m_in, _ = find_fixed_points(Delta, rho, max_iter=max_iter, tol=tol)
        if abs(m_un - m_in) < precision:
            low = mid
        else:
            high = mid
    Delta_Alg = Delta
    # Find Delta_Dyn in [mid, right] 
    low, high = mid, right
    while high - low > 1e-4:
        mid = (low + high) / 2
        Delta = mid * rho**2
        m_un, m_in, _ = find_fixed_points(Delta, rho, max_iter=max_iter, tol=tol)
        if abs(m_un - m_in) < precision:
            high = mid
        else:
            low = mid
    Delta_Dyn = Delta
    # Find Delta_IT in [Delta_Alg, Delta_Dyn]
    low, high = Delta_Alg/rho**2, Delta_Dyn/rho**2
    while high - low > 1e-4:
        mid = (low + high) / 2
        Delta = mid * rho**2
        m_un, m_in, _ = find_fixed_points(Delta, rho, max_iter=max_iter, tol=tol)
        Phi_un, Phi_in = [Phi_RS(m, Delta, rho) for m in [m_un, m_in]]
        if Phi_un < Phi_in:
            low = mid
        else:
            high = mid
    Delta_IT = Delta
    return Delta_Alg, Delta_IT, Delta_Dyn
```

```python
rho, Delta = 0.075, np.linspace(0.95, 1.4, 100)
m_un, m_in, m_unstable = [np.zeros_like(Delta) for _ in range(3)]
Phi_un, Phi_in, Phi_unstable = [np.zeros_like(Delta) for _ in range(3)]
for i, d in log_progress(list(enumerate(Delta)), every=1):
    m_un[i], m_in[i], m_unstable[i] = find_fixed_points(d,rho)
    Phi_un[i], Phi_in[i], Phi_unstable[i] = [Phi_RS(m,d,rho) for m in [m_un[i], m_in[i], m_unstable[i]]]
Delta_Alg, Delta_IT, Delta_Dyn = find_Delta_threshold((0.95/rho**2, 1.2/rho**2, 1.4/rho**2), rho)
```

```python
plt.figure(figsize=(10,5))
idx_left, idx_right = 10, -7
plt.plot(Delta[:idx_right], m_in[:idx_right], color='blue', label='stable branch')
plt.plot(Delta[idx_left:], m_un[idx_left:], color='blue')
plt.plot(Delta[idx_left:idx_right], m_unstable[idx_left:idx_right], color='red', label='unstable branch')
plt.axvline(x=Delta_Alg, color='red', linestyle='-.', label=r'$\Delta_{\mathrm{Alg}}$')
plt.axvline(x=Delta_Dyn, color='red', linestyle='--', label=r'$\Delta_{\mathrm{Dyn}}$')
plt.axvline(x=Delta_IT, color='black', linestyle='--', label=r'$\Delta_{\mathrm{IT}}$')
plt.xlabel(r'$\Delta$')
plt.ylabel(r'$m^*$')
plt.title(r'Two balanced groups $\rho={0}$'.format(rho))
plt.legend()
plt.savefig('hw4_5_1.pdf', bbox_inches='tight')
```
